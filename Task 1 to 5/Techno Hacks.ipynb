# **Reading File**

```python
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/Dataset/Employee Sample Data.csv', encoding= 'unicode_escape')
data.head(25)
```

# **Removing Missing Values and Removing Outliers**

```python
# Assuming your DataFrame is named 'data'
data['Exit Date'].fillna('Ongoing', inplace=True)

from scipy import stats
import numpy as np

# Define a Z-score threshold (e.g., 3)
z_threshold = 3

# Identify and remove outliers in 'Age'
data = data[(np.abs(stats.zscore(data['Age'])) < z_threshold)]

# Identify and remove outliers in 'Annual Salary'
data = data[(np.abs(stats.zscore(data['Annual Salary'].str.replace('[\$,]', '', regex=True).astype(float))) < z_threshold)]

# Display the cleaned DataFrame
data.head(25)
```

# **Summary Statistics**

```python

# Display summary statistics for the entire DataFrame
summary_stats = data.describe()

# Calculate mode separately
mode_values = data.mode().iloc[0]

# Display the mean, median, mode, and standard deviation
mean_value = data.mean()
median_value = data.median()
std_deviation = data.std()

print("Mean:")
print(mean_value)

print("\nMedian:")
print(median_value)

print("\nMode:")
print(mode_values)

print("\nStandard Deviation:")
print(std_deviation)
```

# **Visualization using Histogram**


```python
import pandas as pd
import matplotlib.pyplot as plt


# Extract the 'Age' column
age_data = data['Age']

# Create a histogram
plt.figure(figsize=(10, 6))
plt.hist(age_data, bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

# **Pivot Tables**

```python
import pandas as pd


# Remove dollar signs and commas from 'Annual Salary' column and convert to numeric
data['Annual Salary'] = pd.to_numeric(data['Annual Salary'].replace('[\$,]', '', regex=True), errors='coerce')

# Create a pivot table to summarize data
pivot_table = pd.pivot_table(data, values='Annual Salary', index='Gender', columns='Department', aggfunc='mean', margins=True, margins_name='Total')

# Display the pivot table
print("Pivot Table:")
print(pivot_table)
```

# **Removing Duplication**

```python
import pandas as pd

# Identify duplicate rows
duplicate_rows = data[data.duplicated()]

# Display duplicate rows
print("Duplicate Rows:")
print(duplicate_rows)

# Remove duplicate rows and create a new DataFrame
data_no_duplicates = data.drop_duplicates()

# Display the DataFrame after removing duplicates
print("\nDataFrame after Removing Duplicates:")
data.tail(20)
```
